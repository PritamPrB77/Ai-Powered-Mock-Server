```python
import requests

OLLAMA_URL = "http://localhost:11434/api/generate"
MODEL_NAME = "mistral"   # Change to "qwen2.5:3b" if you prefer

def chat_with_ollama(prompt):
    payload = {
        "model": MODEL_NAME,
        "prompt": prompt,
        "temperature": 0.2,
        "stream": False
    }

    try:
        response = requests.post(OLLAMA_URL, json=payload)
        response.raise_for_status()
        return response.json()["response"]
    except Exception as e:
        return f"Error: {e}"

if __name__ == "__main__":
    print("Local Ollama Chat (type 'exit' to quit)\n")

    while True:
        user_input = input("You: ")

        if user_input.lower() in ["exit", "quit"]:
            print("Exiting...")
            break

        reply = chat_with_ollama(user_input)
        print("\nAI:", reply)
```

```python
import requests

OLLAMA_URL = "http://localhost:11434/api/generate"
MODEL_NAME = "qwen2.5:3b"
# MODEL_NAME = "mistral"

def generate_once(prompt):
    payload = {
        "model": MODEL_NAME,
        "prompt": prompt,
        "temperature": 1.0,
       "num_predict": 60,
        "top_k": 20,
         "top_p": 0.9,
        "stream": False
    }

    response = requests.post(OLLAMA_URL, json=payload)
    return response.json()["response"]

if __name__ == "__main__":
    user_prompt = input("Enter your prompt: ")
    result = generate_once(user_prompt)
    print("\nModel Output:\n")
    print(result)
```
